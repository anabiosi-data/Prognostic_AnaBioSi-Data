```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

# Hyper-params
IMG_HEIGHT, IMG_WIDTH = 300, 400
BATCH_SIZE            = 32
AUTOTUNE              = tf.data.AUTOTUNE

# Preprocess: resize → scale → reduce contrast → clamp
def preprocess(image, numeric, label):
    image = tf.image.resize(image, [IMG_HEIGHT, IMG_WIDTH])
    image = tf.cast(image, tf.float32) / 255.0
    image = tf.image.adjust_contrast(image, 0.5)
    image = tf.clip_by_value(image, 0.0, 1.0)
    return (image, numeric), label


# Augmentation: flips, rotations, shifts, contrast (no random crops/zooms)
data_augment = keras.Sequential([
    layers.RandomFlip("horizontal"),            # Remove vertical if not safe!
    layers.RandomRotation(0.05),                # ±9°
    layers.RandomTranslation(0.05, 0.05),       # 5% shift
    layers.RandomContrast(0.05),                # ±5% contrast
    # Optionally, add Gaussian noise if your images have noise in acquisition
], name="data_augmentation")

def augment(image, numeric, label):
    image = data_augment(image)
    return image, numeric, label



# Build the tf.data pipeline
def make_dataset(images, numerics, labels, shuffle=False, augment_data=False):
    ds = tf.data.Dataset.from_tensor_slices((images, numerics, labels))
    if shuffle:
        ds = ds.shuffle(buffer_size=len(images), seed=42)
    if augment_data:
        ds = ds.map(augment, num_parallel_calls=AUTOTUNE)
    ds = ds.map(preprocess, num_parallel_calls=AUTOTUNE)
    return ds.batch(BATCH_SIZE).prefetch(AUTOTUNE)

# Instantiate train/val/test sets
train_set = make_dataset(
    X_img_train, X_num_train, y_train,
    shuffle=True,  augment_data=True
)
val_set   = make_dataset(
    X_img_val,   X_num_val,   y_val,
    shuffle=False, augment_data=False
)
test_set  = make_dataset(
    X_img_test,  X_num_test,  y_test,
    shuffle=False, augment_data=False
)
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, regularizers, backend as K

# =============================================================================
# CUSTOM LAYER ParametrisedCompatibility
# =============================================================================
@tf.keras.utils.register_keras_serializable()
class ParametrisedCompatibility(layers.Layer):
    def __init__(self, kernel_regularizer=None, **kwargs):
        super().__init__(**kwargs)
        # Allow passing in either a Regularizer instance, a config dict, or a string
        self.kernel_regularizer = regularizers.get(kernel_regularizer)

    def build(self, input_shape):
        C = input_shape[0][-1]
        self.u = self.add_weight(
            name='u',
            shape=(C, 1),
            initializer='uniform',
            regularizer=self.kernel_regularizer,
            trainable=True
        )
        super().build(input_shape)

    def call(self, inputs):
        local, g = inputs
        g_proj = K.expand_dims(K.expand_dims(g, 1), 1)
        combined = tf.nn.tanh(local + g_proj)
        return K.dot(combined, self.u)

    def compute_output_shape(self, input_shape):
        return (input_shape[0][0], input_shape[0][1], input_shape[0][2], 1)

    def get_config(self):
        config = super().get_config()
        # serialize the regularizer (it has its own get_config)
        config.update({
            "kernel_regularizer": regularizers.serialize(self.kernel_regularizer)
        })
        return config
```

```python
# =============================================================================
# MULTI-MODAL ATTENTION MODEL DEFINITION
# =============================================================================
def build_multi_modal_attention(
    img_shape=(300, 400, 3),
    num_num_feats=1,
    num_classes=3,
    reg=0.0005,
    dropout=0.35
):
    # 2.1 Inputs
    img_in = keras.Input(shape=img_shape, name='image_input')
    num_in = keras.Input(shape=(num_num_feats,), name='numeric_input')

    # 2.2 Feature-Extraction Backbone
    x = layers.Conv2D(16, 7, padding='same', kernel_initializer='he_normal')(img_in)
    x = layers.Conv2D(16, 5, padding='same', kernel_initializer='he_normal')(x)
    local12 = layers.Activation('relu')(x)

    x = layers.Conv2D(16, 3, padding='same', kernel_initializer='he_normal')(local12)
    local1 = layers.Activation('relu')(x)

    x = layers.MaxPooling2D(2)(local1)
    x = layers.Dropout(dropout)(x)

    x = layers.Conv2D(32, 3, padding='same', kernel_initializer='he_normal')(x)
    local2 = layers.Activation('relu')(x)

    x = layers.MaxPooling2D(2)(local2)
    x = layers.Dropout(dropout)(x)

    x = layers.Conv2D(64, 3, padding='same', kernel_initializer='he_normal')(x)
    local3 = layers.Activation('relu')(x)

    x = layers.Conv2D(128, 3, padding='same', kernel_initializer='he_normal')(local3)
    x = layers.Activation('relu')(x)

    x = layers.MaxPooling2D(2)(x)
    x = layers.Dropout(dropout)(x)

    # 2.3 Global feature vector g
    flat = layers.Flatten()(x)
    g = layers.Dense(64, activation='relu')(flat)
    g = layers.Dropout(0.2)(g)
    g = layers.Dense(16, activation='relu', name='global_feat')(g)

    # 2.4 Attention branch helper
    def attention_branch(local_feat, branch_name):
        # project local to same dim
        l = layers.Dense(16)(local_feat)
        c = ParametrisedCompatibility(
            kernel_regularizer=regularizers.l2(reg),
            name=f'cpc_{branch_name}'
        )([l, g])
        flat_c = layers.Flatten(name=f'flatc_{branch_name}')(c)
        a = layers.Activation('softmax', name=f'attn_{branch_name}')(flat_c)
        reshaped = layers.Reshape((-1, 16), name=f'reshape_{branch_name}')(l)
        return layers.Lambda(
            lambda z: K.squeeze(
                K.batch_dot(
                    K.expand_dims(z[0], 1),
                    z[1]
                ), 1
            ), name=f'g_{branch_name}'
        )([a, reshaped])

    g1 = attention_branch(local1, 'l1')
    g2 = attention_branch(local2, 'l2')
    g3 = attention_branch(local12, 'l12')

    # 2.5 Numeric branch
    y = layers.Dense(64, activation='relu')(num_in)
    y = layers.Dropout(0.2)(y)

    # 2.6 Fuse and classifier head
    fused = layers.Concatenate(name='fusion')([g3, g1, g2, y])
    z = layers.Dense(64, activation='relu', kernel_regularizer=regularizers.l2(reg))(fused)
    z = layers.Dropout(0.5)(z)
    out = layers.Dense(num_classes, activation='softmax', name='predictions')(z)

    return keras.Model([img_in, num_in], out, name='multi_modal_attention')
```

```python
# =============================================================================
# INSTANTIATE, COMPILE, AND TRAIN
# =============================================================================
# Instantiate
model_att = build_multi_modal_attention(
    img_shape=(300, 400, 3),
    num_num_feats=1,
    num_classes=3
)

# Compile
sgd = keras.optimizers.SGD(learning_rate=0.0035, momentum=0.9, nesterov=True)
model_att.compile(
    optimizer=sgd,
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
model_att.summary()

# Training callbacks
epochs = 100
cosine = tf.keras.optimizers.schedules.CosineDecay(0.0035, epochs)
lr_cb = keras.callbacks.LearningRateScheduler(lambda e, lr: float(cosine(e)), verbose=1)
es_cb = keras.callbacks.EarlyStopping(
    monitor='val_loss',
    patience=20,
    restore_best_weights=True,
    start_from_epoch=20,
    verbose=1
)

import matplotlib.pyplot as plt

# Take the very same train_set you’re about to pass to model.fit(...)
for (img_batch, num_batch), label_batch in train_set.take(1):
    imgs = img_batch.numpy()
    labs = label_batch.numpy()
    break

# Plot a 4×4 grid
plt.figure(figsize=(8,8))
for i in range(16):
    plt.subplot(4,4,i+1)
    plt.imshow(imgs[i])
    plt.title(class_names[int(labs[i])])
    plt.axis('off')
plt.tight_layout()
plt.show()


# Train on existing tf.data pipelines
#    train_set yields ((img, numeric), label)
#    val_set yields ((img, numeric), label)
with tf.device('/device:GPU:0'):
    history = model_att.fit(
        train_set,
        validation_data=val_set,
        epochs=epochs,
        callbacks=[lr_cb, es_cb],
        verbose=2
    )
```

```python
import numpy as np
from sklearn.metrics import confusion_matrix, classification_report

# Evaluate on the test set
loss, acc = model_att.evaluate(
    test_set,
    verbose=1
)
print(f"Test loss: {loss:.4f}")
print(f"Test accuracy: {acc:.4f}")

# Get predicted probabilities and classes
y_test_prob = model_att.predict(test_set, verbose=1)       # shape: (N, 3)
y_test_pred = np.argmax(y_test_prob, axis=1)               # shape: (N,)

# Extract true labels from test_set
y_test_true = np.concatenate([
    labels.numpy().astype(int)
    for (_, _), labels in test_set
], axis=0)

# Sanity-check a few examples
for i in range(5):
    print(
        f"Sample {i}: True={class_names[y_test_true[i]]}, "
        f"Pred={class_names[y_test_pred[i]]}, "
        f"Probs={y_test_prob[i]}"
    )
```

```python
from sklearn.preprocessing import label_binarize
from sklearn.metrics import roc_curve, auc

# Binarize labels
y_bin = label_binarize(y_true_ds, classes=[0,1,2])
n_classes = y_bin.shape[1]

fpr, tpr, roc_auc = {}, {}, {}
for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_bin[:, i], y_score_ds[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

# Macro-average
all_fpr = np.unique(np.concatenate([fpr[i] for i in range(n_classes)]))
mean_tpr = np.zeros_like(all_fpr)
for i in range(n_classes):
    mean_tpr += np.interp(all_fpr, fpr[i], tpr[i])
mean_tpr /= n_classes
fpr["macro"], tpr["macro"] = all_fpr, mean_tpr
roc_auc["macro"] = auc(fpr["macro"], tpr["macro"])

# Plot
plt.figure(figsize=(6,6))
for i, color in zip(range(n_classes), ['aqua','darkorange','cornflowerblue']):
    plt.plot(fpr[i], tpr[i], color=color, lw=2,
             label=f"{class_names[i]} (AUC = {roc_auc[i]:.3f})")

plt.plot(fpr["macro"], tpr["macro"], 'k--', lw=2,
         label=f"Macro avg (AUC = {roc_auc['macro']:.3f})")
plt.plot([0,1], [0,1], 'k--', lw=1)
plt.xlim([0.0,1.0])
plt.ylim([0.0,1.05])
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curves")
plt.legend(loc="lower right")
plt.show()
```
